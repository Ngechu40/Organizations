```python
data_count = df['Country'].value_counts().head(50)

# Create a DataFrame from the 'data_count' Series to access the 'Country' and 'Counts' columns.
df_counts = pd.DataFrame({'Country': data_count.index, 'Counts': data_count.values})

# Sort the DataFrame by the counts in descending order for a more meaningful line plot.
df_counts.sort_values(by='Counts', ascending=False, inplace=True)

# Plot the line plot
plt.figure(figsize=(10, 6))
plt.plot(df_counts['Country'], df_counts['Counts'], marker='o', linestyle='-', color='blue')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.xlabel('Country')
plt.ylabel('Counts')
plt.title('Count of Each Country')
plt.tight_layout()
plt.show()
```
